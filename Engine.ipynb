# Engine
<!-- 2026.03.30 DAJones: created
-->
"Engine" is the name of a *file folder* accessible by Linux on a chromebook.  
It is also the name of a Python *environment* managed by `conda`.
The purpose of "Engine" is to develop the means to reveal and exploit the structure of Clifford algebra.

Topics:
- Elements
- Multivectors
- Ordinal indexing
- Sign Tables
- Sign Table Redundancy
- Reference (Hadamard)
- Subalgebras
- Commutating Bivectors
- Signature and Degeneracy
- Commutating Elements
- Isotrophy
