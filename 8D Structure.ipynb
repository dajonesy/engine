```python
import CliffordTools as CT
import util

dimensions = 5
elements = 1 << dimensions
pseudo_scalar = elements - 1
Grade = [0]
while len(Grade) < elements:
    Grade += [ g+1 for g in Grade ]

#ST = CT.SignTable( list(range(elements)) )
#Ref = [ CT.SignRowToIndex( r ) for r in ST ]

R = list(range(elements))
print( "Xor Test R ", util.XorTest( R ) )
Ref = [ CT.SignRowToIndex( CT.SignRow(r,R) ) for r in R  ]

N = list(range(elements >> 1))
msb = elements >> 1
Parity = [ (1&Grade[n])<<(dimensions-1) for n in N ]
Even = [ n + Parity[n] for n in N ]
Odd = [ Even[n] ^ msb for n in N ]
for n in N:
    even = Even[n]
    odd = msb ^ even
    print( f"{n:>4d} {n:08b}   || {even:>4d} {even:08b}   | {Ref[even]:>4d} {Ref[even]:08b}   || {odd:>4d} {odd:08b}   | {Ref[odd]:>4d} {Ref[odd]:08b}" )

RefEven = [ Ref[even] for even in Even ] 
print( "Xor Test Even ", util.XorTest( Even ) )
print( "Xor Test RefEven ", util.XorTest( RefEven ) )

n = 1
while n < msb:
    even = Even[n]
    odd = msb ^ even
    print( f"{n:>4d} {n:08b}   || {even:>4d} {even:08b}   | {Ref[even]:>4d} {Ref[even]:08b}   || {odd:>4d} {odd:08b}   | {Ref[odd]:>4d} {Ref[odd]:08b}" )
    n <<= 1
```

```python

# the same code as in the previous cell except `SignCol` replaces `SignRow`
# these tables relate to a commutated multiplication rule.

R = list(range(elements))
print( "Xor Test R ", util.XorTest( R ) )
Ref = [ CT.SignRowToIndex( CT.SignCol(r,R) ) for r in R  ]

N = list(range(elements >> 1))
msb = elements >> 1
Parity = [ (1&Grade[n])<<(dimensions-1) for n in N ]
Even = [ n + Parity[n] for n in N ]
Odd = [ Even[n] ^ 1 for n in N ]
for n in N:
    even = Even[n]
    odd = Odd[n]
    print( f"{n:>4d} {n:08b}   || {even:>4d} {even:08b}   | {Ref[even]:>4d} {Ref[even]:08b}   || {odd:>4d} {odd:08b}   | {Ref[odd]:>4d} {Ref[odd]:08b}" )

RefEven = [ Ref[even] for even in Even ] 
print( "Xor Test Even ", util.XorTest( Even ) )
print( "Xor Test RefEven ", util.XorTest( RefEven ) )

n = 1
while n < msb:
    even = Even[n]
    odd = Odd[n]
    print( f"{n:>4d} {n:08b}   || {even:>4d} {even:08b}   | {Ref[even]:>4d} {Ref[even]:08b}   || {odd:>4d} {odd:08b}   | {Ref[odd]:>4d} {Ref[odd]:08b}" )
    n <<= 1
```
