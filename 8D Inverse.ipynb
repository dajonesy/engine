We anticipate a characteristic equation of 16th order.  So, the least number of multivector multiplies would be 4 if we can find a method which doubles the order with each multiply.

I have a strong intuition that the principle elements play a central role.  For the first time in the dimensional progression there are quads that make octos.  (Actually the singular is more appropriate, since there is only the pseudo-scalar octo at this dimension.)  There are 35 ways to split the octo into quads.  We have 7 special splits in the principle elements.  These 14 quads, along with the identity and the pseudo-scalar form a group with a simple inverse method.  That leaves 28 splits, or 56 quads, to account for.
